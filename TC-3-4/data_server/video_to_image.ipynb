```python
import pathlib
import cv2
import os
```

```python
def video2image(video_path):
    ext = os.path.splitext(video_path)[-1]
    video = cv2.VideoCapture(video_path)

    frame_dir = video_path.replace(ext, "")
    os.makedirs(frame_dir, exist_ok=True)

    index = 0
    while True:
        retval, frame = video.read()

        if not retval:
            break

        # print(f"{os.path.dirname(video_path)}/{index:06d}.jpg")
        # print(f"{frame_dir}/{index:06d}.jpg")
        cv2.imwrite(f"{frame_dir}/{index:06d}.jpg", frame)
        index += 1
```

```python
for path in pathlib.Path("/data/kiat/multiple_camera_fall_dataset/dataset").glob("*/*.avi"):
    video2image(str(path))
```
