```python
import cv2
import numpy as np
import matplotlib.pylab as plt
import pathlib
```

```python
def read_video(frame_dir):
    frames = list(map(str, pathlib.Path(f"{frame_dir}").glob("*.jpg")))
    frames.sort()
    return frames

def read_frame(path):
    image = plt.imread(path)
    return image
```

```python
def visualize_video(frame_dir):
    frames = read_video(frame_dir)

    nrows = int(np.ceil(len(frames) / 10))
    ncols = 10
    
    fig, ax = plt.subplots(nrows, ncols, figsize=(2*ncols, 2*nrows))

    for i in range(len(frames)):
        ax[i//10, i%10].imshow(read_frame(frames[i]))
        ax[i//10, i%10].axis("off")

    plt.show()
```

```python
visualize_video("/data/kiat/mpii_videos_flatten/022726963")
```
